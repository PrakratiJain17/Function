```python
#Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even(list1):
  sum=0
  for i in list1:
    if i%2==0:
      sum+=i
  return sum
sum_even([1,2,4,6,5,3,8])
```

```python
#Q3. Implement a Python function that takes a list of integers and returns a new list containing the squares ofeach number.
def square_number(list1):
  result=[]
  for i in list1:
   result.append(i**2)
  return result
square_number([1,3,5,7,9])
```

```python
#Q4.Write a Python function that checks if a given number is prime or not from 1 to 200
def check_prime(n):
  if n<1 or n>200:
    print("Out of range")
  if n<=1:
    print("Not prime")

  for i in range(2,int(n**0.5)+1):
    if n%i == 0:
        print("Not Prime")
        return
  print("Prime Number")

check_prime(5)
check_prime(12)
```

```python
#Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n):
        """Initialize the iterator with the number of terms."""
        self.n = n        # Number of terms in the sequence
        self.a = 0        # First term
        self.b = 1        # Second term
        self.count = 0    # Counter for the number of terms generated

    def __iter__(self):
        """Return the iterator object itself."""
        return self

    def __next__(self):
        """Return the next term in the Fibonacci sequence."""
        if self.count >= self.n:
            raise StopIteration  # Stop iteration when the number of terms is reached

        if self.count == 0:
            self.count += 1
            return self.a

        if self.count == 1:
            self.count += 1
            return self.b

        # Compute the next term
        next_term = self.a + self.b
        self.a = self.b
        self.b = next_term
        self.count += 1
        return next_term
n = 10  # Number of terms in the Fibonacci sequence
fib_iterator = FibonacciIterator(n)

for term in fib_iterator:
    print(term)
```

```python
#Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exp):
  for i in range(exp+1):
    yield 2**i
for j in power_of_two(6):
  print(j)
```

```python
#Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
def sort_tuple(tuple1):
  return sorted(tuple1,key=lambda x:x[1])
sort_tuple([(3,2,),(1,5),(2,1)])
```

```python
#Q9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def conversion(list1):
  return list(map(lambda x:(x*9/5)+32,list1))
conversion([25,100,220,232])
```

```python
#Q10.Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string1):
  vowels='aeiouAEIOU'
  return ''.join(filter(lambda x: x not in vowels,string1))
remove_vowels("HELLO")
```
